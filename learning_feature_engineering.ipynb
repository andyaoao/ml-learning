```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
```

```python
data_path = "Data/train_sample.csv"
```

```python
df_data = pd.read_csv(data_path)
```

```python
df_data.head()
```

```python
# check na value and fillna with mean
df_data.isnull().sum()
df_data = df_data.fillna(df_data.mean())
```

```python
# filter figure columns only
df_figure_only = df_data.drop(["datetime", "Scored"], axis=1)
```

```python
# transform with MinMaxScaler, it will retuen numpy array
minmax_scale = preprocessing.MinMaxScaler()
minmax_scaled_frequency = minmax_scale.fit_transform(df_figure_only)
```

```python
# transform numpy array to pandas dataframe
minmax_scaled_frequency = pd.DataFrame(minmax_scaled_frequency, columns=df_figure_only.columns)
```

```python
# transform with StandardScaler, it will retuen numpy array
standard_scale = preprocessing.StandardScaler()
standard_scaled_frequency = standard_scale.fit_transform(df_figure_only)
```

```python
# transform numpy array to pandas dataframe
standard_scaled_frequency = pd.DataFrame(standard_scaled_frequency, columns=df_figure_only.columns)
```

```python
# transform with RobustScaler, it will retuen numpy array
robust_scale = preprocessing.RobustScaler()
robust_scaled_frequency = robust_scale.fit_transform(df_figure_only)
# transform numpy array to pandas dataframe
robust_scaled_frequency = pd.DataFrame(robust_scaled_frequency, columns=df_figure_only.columns)
```

```python
# create polynomial feature transformor, it will retuen numpy array
polynomial_interaction = PolynomialFeatures(degree=2, include_bias=False)
polynomial_df_figure_only = polynomial_interaction.fit_transform(df_figure_only)
# transform numpy array to pandas dataframe
polynomial_df_figure_only = pd.DataFrame(polynomial_df_figure_only)
```
