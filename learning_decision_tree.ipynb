# Decision Tree

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from IPython.display import Image
from sklearn import tree
import pydotplus
```

```python
iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
df_iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
iris['feature_names']
```

```python
df_iris.head(5)
```

```python
decision_tree = DecisionTreeClassifier(random_state=0)
```

```python
model = decision_tree.fit(df_iris.iloc[:,0:-1], df_iris[["target"]])
```

```python
pred = model.predict([[6, 7, 8, 9]])
```

```python
# merge original df with prediction
df_pred = pd.DataFrame(data=np.c_[df_iris, pred], columns=df_iris.columns.insert(loc=len(df_iris.columns), item="pred"))
```

```python
df_pred.head(5)
```

```python
dot_data = tree.export_graphviz(decision_tree, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
```
